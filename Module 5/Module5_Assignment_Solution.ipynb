### 1. What is Object-Oriented Programming (OOP)?
* OOP is a programming method that uses **objects and classes**. It helps organize code by grouping data and behavior together, making programs easier to understand and manage.


### 2. What is a class in OOP?
* A class is like a **blueprint or template** used to create objects. It defines what data and methods an object will have.



### 3. What is an object in OOP?
* An object is a **real instance of a class**. It represents real-world entities and contains actual values of the class variables.



### 4. What is the difference between abstraction and encapsulation?
- **Abstraction**: Hides unnecessary details and shows only important features.  
- **Encapsulation**: Bundles data and methods together and protects data using access control.


### 5. What are dunder methods in Python?
* Dunder methods are **special built-in methods** like `__init__`, `__str__`, etc. They give special behavior to objects.


### 6. Explain the concept of inheritance in OOP?

### 7. What is polymorphism in OOP?
* Polymorphism means **many forms**. The same method name behaves differently depending on the object calling it.


### 8. How is encapsulation achieved in Python?

### 9. What is a constructor in Python?

### 11.  What is method overloading in Python?

### 12. What is method overriding in OOP?

### 1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print "Bark!".

### 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod # ---> forces all child classes to write their own area().
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r): # constructor of Circle class
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r

class Rectangle(Shape):
    def __init__(self, l, b): # constructor of Rectangle class
        self.l = l
        self.b = b

    def area(self):
        return self.l * self.b

print(Circle(5).area())
print(Rectangle(4,6).area())
```

### 3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.

```python
class Vehicle: # parent class
    def __init__(self, vtype):
        self.type = vtype

class Car(Vehicle): # Child class of Vehicle Class
    pass

class ElectricCar(Car): # Child class of Car and grandchild of Vehicle
    def __init__(self, vtype, battery):
        super().__init__(vtype) # use for call parent constructor
        self.battery = battery

e = ElectricCar("Car", "Lithium")
print(e.type, e.battery)
```

### 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

```python
class BankAccount:
    def __init__(self):
        self.__balance = 0 # if you don't want to give access any variable to object,
                           # you have to put _ _ before variable name
        # in one word now balance act as a Private attribute.

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount

    def check_balance(self):
        return self.__balance

acc = BankAccount() # acc, hold address where BankAccount obj created.
# From here you can't access balance attribute.
acc.deposit(1000)
print(acc.check_balance())
```

### 7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

```python
class MathOperations:
    @classmethod # in classmethod it take class object itself as its 1st argument.
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

print(MathOperations.add_numbers(10,5))
print(MathOperations.subtract_numbers(10,5))
```

### 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
# __add__ also a dunder method use for addition purpose.
    def __add__(self, other): #self and other are both object, we can add them.
        return Vector(self.x + other.x, self.y + other.y)

# __str__ is used for print purpose, it return a string object.
    def __str__(self):
        return f"({self.x},{self.y})"

v1 = Vector(2,3)
v2 = Vector(4,5)
print(v1 + v2)
```

### 14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

```python
class Employee:
    def calculate_salary(self, hours, rate):
        return hours * rate

class Manager(Employee): # inherite from Employee class
    def calculate_salary(self, hours, rate):
        return super().calculate_salary(hours, rate) + 5000 #super() gives power to inherite from its parents..

m = Manager()
print(m.calculate_salary(40,200))
```

### 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod # it create a rule for its child class, they have to put this sound() method at their body.
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa Baa")

Cow().sound()
Sheep().sound()
```

### 18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.

```python
class House:
    def __init__(self, address, price): # here address and price belongs to House class
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, rooms):
        super().__init__(address, price)
        self.number_of_rooms = rooms # number_of_rooms Attributes belongs to Mansion Class

m = Mansion("Kolkata", 5000000, 10)
print(m.address, m.price, m.number_of_rooms)
```
