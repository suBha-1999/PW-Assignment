## 1. What are data structures, and why are they important?
* Data structures are organized ways to store, manage, and retrieve data efficiently in a computer.

* Data Structures important because:
    1. Efficient Data Management
    2. Improve Performance(like speed, memory use and cpu work)
    3. Backbone of Algorithms(sorting, searching, pathfinding, machine learning)
    4. Better Program Design

## 2.Explain the difference between mutable and immutable data types with examples.
* Mutable data types can be changed after creation (e.g., list, dictionary).
* Immutable data types cannot be changed once created (e.g., string, tuple, int).


## 3.What are the main differences between lists and tuples in Python?
* Lists are mutable, slower, and use [] for list creation.
* Tuples are immutable, faster, and use () for tuples creation.

## 4.Describe how dictionaries store data.
* Dictionaries store data as key-value pairs using a hash table, It allows fast lookups by key.

## 5.Why might you use a set instead of a list in python?
* Sets automatically remove duplicates and allow faster searching element than lists.

## 6.What is a string in python, and how it is different from a list?
* A string is an immutable sequence of characters, but a list is a mutable sequence of items.

## 7.How do tuples ensure data integrity in Python?
* Tuples are immutable, so their values cannot be changed accidentally, ensuring data remains constant.

## 8.What is a hash table, and how does it relate to dictionaries in Python?
* A hash table stores data using a computed hash index.
* Python dictionaries use hash tables to map keys to values quickly.

## 9.Can lists contain different data types in Python?
* Yes. A list can store mixed data types such as integers, strings, floats, or even other lists.

## 10.Explain why strings are immutable in Python?
* Strings are immutable for memory efficiency, security, and to allow safe sharing between programs.

## 11.What advantages do dictionaries offer over lists for certain tasks?
* Dictionaries provide fast access using keys and are ideal for structured data, unlike lists which use indexes.

## 12.Describe a scenario where using a tuple would be preferable over a list.
* Use a tuple when you need fixed data, like coordinates (x, y) or constant settings, to prevent modification.

## 13.How do sets handle duplicate values in Python?
* Sets automatically remove duplicates, storing only unique elements.

## 14.How does the “in” keyword work differently for lists and dictionaries?
* For lists, in checks for values.
For dictionaries, in checks for keys, not values.

## 15.Can you modify the elements of a tuple? Explain why or why not?
* No. Tuples are immutable, meaning their elements cannot be changed after creation.

## 16.What is a nested dictionary, and give an example of its use case?
* A nested dictionary is a dictionary inside another dictionary, useful for storing structured data like student → marks.

  Example: {"student": {"math": 90, "science": 85}}.

## 17.Describe the time complexity of accessing elements in a dictionary?
* Dictionary access is O(1) on average because of hash table indexing.

## 18.In what situations are lists preferred over dictionaries?
* Lists are preferred when data is ordered, indexed, or when duplicates are allowed.

## 19.Why are dictionaries considered unordered, and how does that affect data retrieval?
* Dictionaries do not store elements by position; they store by key hash, so you cannot rely on index-based order.

## 20.Explain the difference between a list and a dictionary in terms of data retrieval.
* Lists retrieve data using indexes, while dictionaries retrieve data using keys, making dictionary lookups faster.
