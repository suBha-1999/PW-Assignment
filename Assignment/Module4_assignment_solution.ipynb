# 1. What is the difference between a function and a method in Python?

## Function

* A block of code that performs a specific task.

* Defined using def keyword.

* Can be called independently.
### Example:
```
def add(a, b):
    return a + b

print(add(5, 3))
```

## Method

* A function that belongs to an object or class.

* Called using object → object.method()
### Example:
```
text = "hello"
print(text.upper())   # upper() is a string method use for uppercase all letter.
```

#  10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Function     |       	      Purpose             |	      Example
-------------|----------------------------------|---------------------
map()	       | Applies function to all items    | map(f, list)
-------------|----------------------------------|---------------------
filter()	   | Keeps items that return True	    | filter(f, list)
-------------|----------------------------------|---------------------
reduce()	   | Reduces list into a single value	| reduce(f, list)
-------------|----------------------------------|---------------------

### Example:
```
from functools import reduce

nums = [1, 2, 3, 4]

print(list(map(lambda x: x*x, nums)))       # map
print(list(filter(lambda x: x%2==0, nums))) # filter
print(reduce(lambda a,b: a+b, nums))        # reduce
```

##  1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

````python
# 1st process
def list_sum(List):
  sum = 0
  for i in range(len(List)): # here we fatch element using index
    if List[i]%2 == 0:
      sum += List[i]
  return sum

# 2nd Process
def list_sum2(List):
  sum = 0
  for num in List: # Here we directly fatch element
    if num % 2 == 0:
      sum += num
  return sum

# main code
List_of_number = [23,10,36,45,92]
print("Sum of all even Number using 1st func: ",list_sum(List_of_number))

print("Sum of all even Number using 2nd func: ",list_sum2(List_of_number))
````

##  2. Create a Python function that accepts a string and returns the reverse of that string.


````python
def rev_string(word):
  return word[len(word)::-1] # after Slicing operation we access element from last to first
  # word[::-1] ----> We can also write like

# Main Code
inp_str = "Subhajit"
print(rev_string(inp_str))
````

## 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

````python
def sqr_num(List):
  sqr_list = [] # take empty list for store square
  for num in List: # access direct element
    sqr_list.append(num**2) # append square one by one in new list
  return sqr_list

# Main Code
num_list = [1, 2, 3, 4, 5]
print(sqr_num(num_list))
````

##  4. Write a Python function that checks if a given number is prime or not from 1 to 200.


````python
# Numbers ≤ 1 are not prime.
# A prime number has no divisors other than 1 and itself.

def is_prime(num):
  if num <= 1:
    return False

  for i in range(2, num):
    if num % i == 0: # if num is divisible by i, num definatly cant be a prime number.
    # All number divisible by 0 and 1.
    # if the number divisible from 2 to till that particular number, we can find number prime or not.
      return False

  return True

# Main Code
prime_list = []
not_Prime_List = []
for n in range(1, 201):
  if is_prime(n):
    prime_list.append(n)
  else:
    not_Prime_List.append(n)

print("Prime number are: ", prime_list)
print("Not prime number are: ", not_Prime_List)
````

#  5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

````python
def fib(n):
  f = lambda n : n if n <= 1 else fib(n-1) + fib(n-2)
  return f(n)

num = int(input("Enter a number: "))
for i in range(num):
  print(fib(i))
````

#  6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


````python
# Function Declaration
def gen(num):
  for i in range(1, num+1):
    yield 2**i

# Main Code
num = int(input("Enter your number: "))
for n in gen(num):
  print(n)
````

#  7. Implement a generator function that reads a file line by line and yields each line as a string.


````python
def read_file_lines(filename):
    with open(filename, 'r') as file: # file open in read mode, only can read
        for line in file:
            yield line.rstrip("\n") # rstrip break lines while got \n

# Main Code
for line in read_file_lines("/content/Assignment for Module4.txt"):
    print(line)
````

#  9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


````python
celsius = [0, 10, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print(fahrenheit)
````

#  10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

````python
text = input("Enter a string: ")

vowels = "aeiouAEIOU"

result = "".join(filter(lambda ch: ch not in vowels, text))

print("String without vowels:", result)
````
